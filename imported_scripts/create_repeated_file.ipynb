```python
import numpy as np

# -------- USER SETTINGS --------
input_file = "./waveforms/100k_qpsk_samples.dat"
output_file = "./waveforms/repeat_qpsk.dat"
sample_rate     = 10_000_000      # Hz
repeat_duration = 0.1             # seconds (start repeating every 0.1s)
total_duration  = 5.0             # total output file length in seconds


# -------- LOAD INPUT SAMPLES --------
samples = np.fromfile(input_file, dtype=np.int16)
num_in = len(samples)

# -------- COMPUTE SIZES --------
samples_per_repeat = int(sample_rate * repeat_duration)
total_samples      = int(sample_rate * total_duration)

# Allocate output buffer (silence)
out = np.zeros(total_samples, dtype=samples.dtype)

# -------- PERIODICALLY INSERT SAMPLES --------
pos = 0
while pos < total_samples:
    # place the input block at the start of this repeat window
    end = pos + num_in
    out[pos:end] = samples[:max(0, min(num_in, total_samples - pos))]
    pos += samples_per_repeat


# -------- SAVE OUTPUT --------
out.tofile(output_file)

print("Done.")
print("Output samples:", len(out))
print("Total duration seconds:", total_duration)
```
